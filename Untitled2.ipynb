```python
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from urllib.request import urlopen
```

```python
url = "https://www.fortishealthcare.com/india/doctors"
html = urlopen(url)
soup = BeautifulSoup(html, 'lxml')
type(soup)
```

```python
soup.title
soup.title.string
soup.a
```

```python
all_links=soup.find_all('a')
for link in all_links:
    print (link.get("href"))
```

```python
all_tables=soup.find_all('table')
right_table=soup.find('table', {"class_":"list_of_doctors"})
```

```python
#Generate lists
A=[]
B=[]
C=[]
D=[]
E=[]
for row in right_table("tr"):
    cells = row.find_all('td')
    states=row.find_all('th') #To store second column data
    if len(cells)==4: #Only extract table body not heading
        A.append(cells[0].find(text=True))
        B.append(states[0].find(text=True))
        C.append(cells[1].find(text=True))
        D.append(cells[2].find(text=True))
        E.append(cells[3].find(text=True))
```

```python
df=pd.DataFrame(A,columns=['Number'])
df['name_of_doctor']=B
df['post_of_doctor']=C
df['specialization']=D
df['departments']=E
df
```
